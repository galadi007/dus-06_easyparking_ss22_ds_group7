In this notebook we will merge files from different json and csv files about parking in Cologne.

```python
import pandas as pd
import numpy as np
import json
```

###  Disabled parking data

```python
def parse_features(features: list):
    return [{**feature['attributes'], **feature['geometry']} for feature in features]
```

```python
with open("disabled.json", "r", encoding='utf8') as fp:
    disabled_parking = json.load(fp)

df_disabled_parking = pd.DataFrame.from_dict(parse_features(disabled_parking['features']))
df_disabled_parking['type'] = 'disabled parking slot'
df_disabled_parking
```

```python
df_disabled_parking.to_csv("Behindertenparkplatzliste.csv")
```

## Parking Garage scraped Data

```python
with open("parkhaus.json", "r") as fp:
    parking_garages = json.load(fp)

df_parking_garages = pd.DataFrame.from_dict(parse_features(parking_garages['features']))
df_parking_garages['type'] = 'parking garage'
df_parking_garages
```

```python
df_parking_garages.to_csv("Parkhausliste_offene_Daten_Köln.csv")
```

## Park and Ride Data

```python
with open("park_and_ride.json", "r", encoding='utf-8') as fp:
    df_park_and_ride = json.load(fp)

df_park_and_ride = pd.DataFrame.from_dict(parse_features(df_park_and_ride['features']))
df_park_and_ride['type']='park and ride'
df_park_and_ride
```

## Parking Meter Data

```python
df_parking_meter = pd.read_csv("parking_meter_2020.csv", sep=';')
df_parking_meter
```

We have to rename GeoKoordinateNord and GeoKoordinateOst to y and x respectively to aling them with other data we have

```python
df_parking_meter = df_parking_meter.rename(columns={'GeoKoordinateOst': 'x', 'GeoKoordinateNord': 'y'})
df_parking_meter['type']='parking meter'
df_parking_meter.head()
```

x and y have comma instead of a dot. We need to replace the commas with dots to parse floats correctly.

```python
df_parking_meter[['x','y']]
```

lets remove the rows which don't have a x,y coordinate

```python
df_parking_meter.dropna(subset=['x','y'], inplace=True)
```

```python
df_parking_meter[['x', 'y']].applymap(lambda x: float(x.replace(',', '.'))).head()
```

```python
df_parking_meter[['x', 'y']] = df_parking_meter[['x', 'y']].applymap(lambda x: float(x.replace(',', '.')))
```

## Charging Station Data

chargin_station.csv had a header consisting of 10 rows which was removed manually in libreoffice to make it easy to read with pandas 

```python
df_charging_station = pd.read_csv('charging_station_edited.csv', sep=';', encoding='latin1')
df_charging_station.iloc[0]
```

```python
df_charging_station = df_charging_station.rename(columns={'Breitengrad':'y', 'Längengrad':'x'})
df_charging_station['type'] = 'charging station'
df_charging_station.iloc[1]
```

### Merge all of them to one dataframe

```python
df_merged = df_disabled_parking.append(df_park_and_ride).append(df_parking_meter).append(df_parking_garages).append(df_charging_station)
df_merged.reset_index(drop=True, inplace=True) 
df_merged
```

```python
df_merged.dropna(subset=['x','y'], inplace=True)
df_merged.to_csv('merged_data.csv')
```
